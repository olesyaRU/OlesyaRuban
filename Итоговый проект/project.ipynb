### Первый датасет

```python
import pandas as pd
data1 = pd.read_csv('diabetes_dataset/diabetes.csv', sep=',')
print(data1)
```

```python
display(data1.head())
```

```python
# Клиническая характеристика пациентов по всей выборке
data1.describe()
```

```python
# Уточним данные по типу показателей: количественные, категориальные
data1.info()
```

```python
# Сгруппируем данные по показателю "Исход", оценим среднее арифметическое и стандартное отклонение
data1_mean = data1.groupby(data1['Outcome']).mean()
data1_std = data1.groupby(data1['Outcome']).std()
print(data1_mean)
print(data1_std)
# Как видно, пациенты с СД имеют больше беременностей в анамнезе, ИМТ, старшего возраста, толщина кожной складки. У них выше глюкоза, инсулин, артериальное давление.
```

```python
# Сгруппируем данные по показателю "Исход", оценим медиану, 25 и 75 процентили
data1_median = data1.groupby(data1['Outcome']).median()
data1_quantile1 = data1.groupby(data1['Outcome']).quantile(0.25)
data1_quantile2 = data1.groupby(data1['Outcome']).quantile(0.75)
print(data1_median)
print(data1_quantile1)
print(data1_quantile2)
# Как видно, пациенты с СД имеют больше беременностей в анамнезе, ИМТ, старшего возраста, толщина кожной складки. У них выше глюкоза, инсулин, артериальное давление.
```

```python
import seaborn as sns
# Показать с помощью графика "ящик с усами" распределение возраста в двух группах: с сахарным диабетом (СД) и без заболевания
fig = plt.figure(figsize=(5, 3))
boxplot = sns.boxplot(
    data=data1,
    y='Outcome',
    x='Age',
    orient='h',
    width=0.9
)
boxplot.set_title('Распределение исходов по возрасту');
boxplot.set_xlabel('Возраст');
boxplot.set_ylabel('Исход');
boxplot.grid()
# Ка видно из графика, пациенты с СД были старше. 
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 6))
sns.histplot(
    data=data1,
    x='Age',
    bins=15,
    kde=True,
    ax=axes[0]
);
axes[0].set_title('Гистограмма распределения исходов по возрасту', fontsize=12)
sns.histplot(
    data=data1,
    x='Age',
    y='Outcome',
    bins=15,
    color='red',
    ax=axes[1]
);
```

```python
# Поскольку гистограммы не являются наглядным отражением распространенности признака, все последующие параметры представлены в виде box plot.
```

```python
# Показать с помощью графика "ящик с усами" распределение беременностей в двух группах: с сахарным диабетом (СД) и без заболевания
fig = plt.figure(figsize=(5, 3))
boxplot = sns.boxplot(
    data=data1,
    y='Outcome',
    x='Pregnancies',
    orient='h',
    width=0.9
)
boxplot.set_title('Распределение исходов по количеству беременностей в анамнезе');
boxplot.set_xlabel('Предшествующие беременности');
boxplot.set_ylabel('Исход');
boxplot.grid()
# Ка видно из графика, пациенты с СД имели больше беременностей в анамнезе. 
```

```python
# Показать с помощью графика "ящик с усами" распределение ИМТ в двух группах: с сахарным диабетом (СД) и без заболевания
fig = plt.figure(figsize=(5, 3))
boxplot = sns.boxplot(
    data=data1,
    y='Outcome',
    x='BMI',
    orient='h',
    width=0.9
)
boxplot.set_title('Распределение исходов по ИМТ');
boxplot.set_xlabel('Индекс массы тела');
boxplot.set_ylabel('Исход');
boxplot.grid()
# Ка видно из графика, пациенты с СД имели больше ИМТ. 
```

```python
# Показать с помощью графика "ящик с усами" распределение АД в двух группах: с сахарным диабетом (СД) и без заболевания
fig = plt.figure(figsize=(5, 3))
boxplot = sns.boxplot(
    data=data1,
    y='Outcome',
    x='BloodPressure',
    orient='h',
    width=0.9
)
boxplot.set_title('Распределение исходов по АД');
boxplot.set_xlabel('Артериальное давление');
boxplot.set_ylabel('Исход');
boxplot.grid()
# Ка видно из графика, пациенты с СД имели выше АД. 
```

```python
# Показать с помощью графика "ящик с усами" распределение толщины кожной складки в двух группах: с сахарным диабетом (СД) и без заболевания
fig = plt.figure(figsize=(5, 3))
boxplot = sns.boxplot(
    data=data1,
    y='Outcome',
    x='SkinThickness',
    orient='h',
    width=0.9
)
boxplot.set_title('Распределение исходов по толщине кожной складки');
boxplot.set_xlabel('Толщина кожной складки');
boxplot.set_ylabel('Исход');
boxplot.grid()
# Ка видно из графика, пациенты с СД имели больше толщину кожной складки. 
```
