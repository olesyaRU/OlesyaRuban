### Чтобы использовать несколько ограничений, можно использовать логические и &, или |. 
Например:                                                                                                
data = data.loc[(data['trestbps'] != 0)& (data['chol'] != 0)]
print(data)
Запишет в data только строки в которых одновременно выполняются два условия.
